# Read Dataset

```python
import pandas as pd
df_train = pd.read_csv('./Input/training_set.csv')
df_train.head()
```

# Drop ID column as it is Statistically Insignificant

```python
df_train = df_train.drop(labels=['Loan_ID'],axis=1)
```

# Converting Credit History to object

```python
crd_hst = []
for i in range(len(df_train)):
    if df_train['Credit_History'][i]==1:
        crd_hst.append('Yes')
    elif df_train['Credit_History'][i]==0:
        crd_hst.append('No')
    else:
        crd_hst.append(None)
df_train['Credit_History']=crd_hst
```

# Pycaret Setup

```python
from pycaret.classification import setup, compare_models, pull
setup(df_train,target='Loan_Status')
setup_df = pull()
```

# Compare best models

```python
best_model = compare_models()
compare_df = pull()
```
